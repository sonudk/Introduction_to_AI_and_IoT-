### Loading libraries and data

```python
from sklearn.datasets import load_iris
import numpy as np

np.random.seed(10) # For random weight initialization

iris = load_iris() # iris is a python dictionary with key-value pairs
```

#### Pre-process Data
Convert labels to categorical - One-hot encoding

```python
X = iris['data']
Y = iris['target']
from keras.utils import to_categorical

Ny = len(np.unique(Y)) # Ny is number of categories/classes


Y = to_categorical(Y[:], num_classes = Ny) # 
```

#### Train - Test split
#### Data Normalization/Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 1)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train) # Computes the mean and standard deviation

X_train = scaler.transform(X_train) # Perform transformation: x = (x-mean)/std
X_test = scaler.transform(X_test)
```

### Keras

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(Ny, activation='softmax')) # 'sigmoid'
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # 'binary_crossentropy'
    # 'adam', 'adagrad', 'sgd'
```

```python
model.summary()
```

```python
# Fit the model
model.fit(X_train, Y_train, epochs=500, batch_size=100)
```

```python
# calculate predictions
predictions = model.predict(X_train)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

```python
# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
predictions = np.argmax(model.predict(X_test), axis=1)
Y_test = np.argmax(Y_test,axis=1)
print(confusion_matrix(Y_test,predictions))  
print(classification_report(Y_test,predictions))  
```
